```python
from transformers import pipeline
```

# Sentiment Analysis

```python
# create text classification pipeline, since i didn't mention any specific model it use default one
text_classifer=pipeline(task="text-classification")
```

```python
text="I was happy"
result=text_classifer(text)
print(result)
```

```python
text="It's very bad"
result=text_classifer(text)
print(result)
```

```python
Model_url="https://huggingface.co/cardiffnlp/twitter-roberta-base-sentiment"
```

```python
#pipeline with choosen model
text_classifer=pipeline(task="text-classification",model="cardiffnlp/twitter-roberta-base-sentiment")
```

```python
text="I'm happy"
result=text_classifer(text)
print(result)
```

```python
text="I'm not happy"
result=text_classifer(text)
print(result)
```

# Text Summerization

```python
Model_url="https://huggingface.co/Falconsai/text_summarization"
```

```python
#initializes a summerization pipeline
summerizer=pipeline(task="summarization", model="Falconsai/text_summarization")
```

```python
text="""Technology has transformed the way we live, work, and communicate. From smartphones to artificial intelligence, it shapes nearly every aspect of modern life. It enables instant access to information, enhances productivity, and connects people across the globe. Education, healthcare, and business have all benefited from technological advancements, improving efficiency and outcomes. However, these changes also raise concerns, including privacy issues, job displacement, and social isolation. As technology continues to evolve, it’s crucial to strike a balance between innovation and ethical responsibility. Embracing progress while considering its broader impact ensures a more inclusive and sustainable future for everyone.
"""
result=summerizer(text,min_length=10,max_length=50)
print(result)
```
